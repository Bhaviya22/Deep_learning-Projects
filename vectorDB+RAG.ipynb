```python
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss
```

```python
sentence=[
    "i love cats",
    "Dogs are the best",
    "Artificial Intelligence is future",
    "cats are cute animals",
    "i enjoy machine learning",
    "i have a pet dog"

]
#load pre tarined sentence embedding model
model=SentenceTransformer("all-MiniLM-L6-v2")
```

```python
embeddings=model.encode(sentence)    #shape(6,384) here 6 is the no of sentences and 384 is way of multidimensional vector of each sentence
embeddings=np.array(embeddings).astype("float32")
```

```python
embeddings.shape
```

**384->will have index and by using eucledian distance**

```python
dimension=embeddings.shape[1]
index=faiss.IndexFlatL2(dimension) # This creates a flat (non-hierarchical) index that uses L2 distance(eucledian)
index.add(embeddings)  #This adds your embedding vectors to the FAISS index so you can search for similar vectors later.
```

```python
query="i adore kittens"
query_vector=model.encode([query]).astype("float32")
top_k=3 #we need top 3 sentences
distances,indices=index.search(query_vector,top_k) #indices we have sentence
print("query:",query)
print("\nTop matches")
for i in indices[0]:  #[0] query index
  print("-",sentence[i])
```
