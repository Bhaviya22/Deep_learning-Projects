```python
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss
import google.generativeai as genai
```

```python
# Configure Gemini with your API key
genai.configure(api_key="")
```

```python
# Example: your sentence list
sentences = [
    "Artificial Intelligence is the field of making machines intelligent.",
    "Machine learning is a subset of AI that enables systems to learn from data.",
    "Deep learning uses neural networks to model complex patterns in data.",
    "The Eiffel Tower is in Paris.",
    "Supervised learning uses labeled datasets."
]
```

```python
# Generate embeddings
embedder = SentenceTransformer('all-MiniLM-L6-v2')
sentence_embeddings = embedder.encode(sentences).astype("float32")
```

```python
# Build FAISS index
index = faiss.IndexFlatL2(sentence_embeddings.shape[1])
index.add(sentence_embeddings)
```

```python
query = "How do machines learn?"
query_embedding = embedder.encode([query]).astype("float32")

# Search for similar sentences
k = 3
distances, indices = index.search(query_embedding, k)
print("query:",query)
print("\nTop matches")
for i in indices[0]:  #[0] query index
  print("-",sentences[i])
```
