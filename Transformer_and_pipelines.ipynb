```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
classifier("This is the best day of my life")
```

```python
classifier=pipeline("zero-shot-classification")
classifier("I am working as teacher ",
           candidate_labels=["eduaction","bussiness"])
```

```python
generator=pipeline("text-generation")  #
generator("In this course,we going to")
```

```python
pipe = pipeline("text-generation", model="K-intelligence/Midm-2.0-Base-Instruct")
pipe("there was a girl lived in a city")
```

```python
ummask=pipeline("fill-mask")
ummask("the class students are <mask>",top_k=2)
```

```python
ner=pipeline("ner",grouped_entities=True)
ner("My name is bav and i work at hugging face in brooklyn")
```
