```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset=sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
df = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
df['label']=breast_cancer_dataset.target
```

```python
x=df.drop('label',axis=1)
y=df['label']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

# Fit on training data and transform
x_train_std = sc.fit_transform(x_train)

# Transform test data using the same scaler
x_test_std = sc.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow import keras
```

**1.flattern -to convert 1 dimensional vector**


**2.dense-to get good connectivity among nodes**


**3.optimizer- if we get high biass means allow to change to the weights**



```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history=model.fit(x_train_std,y_train,validation_split=0.1,epochs=50)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(['training data','validation'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model accuracy')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['training data','validation'],loc='lower right')
```

```python
y_predict =model.predict(x_test_std)
print(y_predict)
```

```python
y_predict1=np.argmax(y_predict,axis=1)
print(y_predict1)
```
